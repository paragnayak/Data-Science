## Binning Numerical Data (Age)
The Age column is a continuous numerical variable. We can bin it into categorical groups like "Child," "Adult," and "Senior" to capture non-linear relationships. We'll use cut() for this.

```julia
@time begin

using Pkg
Pkg.add("CategoricalArrays") # You only need to do this once

using CategoricalArrays
using DataFrames

# Now the 'cut' function is available
titanic_df.Age_Group = cut(titanic_df.Age, [0, 12, 60, Inf], labels=["Child", "Adult", "Senior"])
end
```

## Creating a Family Size Feature
The SibSp (siblings/spouses) and Parch (parents/children) columns can be combined to create a single, more informative feature: FamilySize.

```julia
@time begin

# Create a 'FamilySize' column
# Add 1 to include the person themselves
titanic_df.FamilySize = titanic_df.SibSp + titanic_df.Parch .+ 1

end
```

## Creating a 'IsAlone' Feature
Based on the new FamilySize column, we can create a binary feature IsAlone, which is often a strong predictor of survival.

```julia
@time begin
    
# Create an 'IsAlone' column (true if FamilySize is 1)
titanic_df.IsAlone = titanic_df.FamilySize .== 1

end
```

## Handling Missing Values (Fare)
Although not strictly feature engineering, dealing with missing values is a crucial part of preparing data. The Fare column has a few missing values. We can impute them with the median or mean.

```julia
@time begin

using Statistics

# Impute missing 'Fare' values with the median
median_fare = median(skipmissing(titanic_df.Fare))
replace!(titanic_df.Fare, missing => median_fare)

end
```

MLJ.jl uses a common fit! and predict workflow. The key difference is the concept of a "machine," which binds the model to the data.

```julia
@time begin

using MLJ, DecisionTree, DataFrames

# Assume X and y are loaded from the Titanic dataset
# Ensure you have already handled `missing` values

# Coerce the feature data types for MLJ
# `Sex` is categorical, so it needs to be Multiclass
# `Pclass` is a number but represents categories, so it should be OrderedFactor
X = coerce(X, :Sex => Multiclass, :Pclass => OrderedFactor)

# Coerce the target variable to the correct type
# This is crucial for MLJ to treat it as a classification problem
y = coerce(y, Multiclass)

# Now, split the data
X_train, X_test = partition(X, 0.8, shuffle=true, rng=123)
y_train, y_test = partition(y, 0.8, shuffle=true, rng=123)

# Load the model
model = DecisionTreeClassifier()

# Now, `machine` will work with the correctly typed data
mach = machine(model, X_train, y_train)

# Fit and predict as before
fit!(mach)
y_pred = predict(mach, X_test)

end
```
