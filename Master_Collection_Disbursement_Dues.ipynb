```
# Removing the Non existent value columns from the respective dataset
df1=df1.drop('Source Name',axis=1)
```

```
# Removing the Non existent value columns from the respective dataset
df2=df2.drop(['Sub Source Name','Collection Executive','Average DPD'],axis=1)
```

```
df_1=df1[['Disbursement Date','Average DPD']]
```

```
# Merging multiple data frames into one data frame
df_final=df2.append(df_1)
```

```
# Checking the first few rows of the data
df_final.head()
```

```
df_final.info()
```

```
# Taking Only the Three Columns Disbursement Date, Source Name and Current DPD
df_final1=df_final[['Disbursement Date','Source Name','Current DPD', 'Average DPD']]
df_final1.head(20)
```

```
df_final1=df_final1.rename(columns={'Average DPD':'Average_DPD'})
df_final1=df_final1.rename(columns={'Current DPD':'Current_DPD'})
```

```
df_final1['Current_DPD'].fillna(value=0, inplace=True)
```

```
#Defining the function to differentiate among the dues
def bin_dffinal(x):
    
    if(x>=30 and x<60):
        return('30+')
    elif(x>=60 and x<90):
        return('60+')
    elif(x>=90):
        return('90+')
    elif(x>1 and x<30):
        return('<30') 
    else:
          return('0')
        

df_final1['Current_DPD_dues'] = df_final1['Current_DPD'].apply(bin_dffinal)
```

```
df_final1.head(20)
```

```
# Dummification of the values of the dues.
dummies = pd.DataFrame(pd.get_dummies(df_final1['Current_DPD_dues']))
dummies.head()
```

```
df_final1 = pd.concat([df_final1, dummies], axis=1)
df_final1.head()
```

```
# . checking the first 10 columns.
df_final1.head(10)
```

```
df_final1.info()
```

```
df_final1['Disbursement Date']= pd.to_datetime(df_final1['Disbursement Date'])
```

```
import datetime as dt
```

```
# Firstly to sum all the values of the respective overdues and then grouping it by the month and year of the disbursement date and the source name.

aggregations = {
    '30+': lambda x:x.sum(),
    '60+': lambda x:x.sum(),
    '90+': lambda x:x.sum(),
    '<30': lambda x:x.sum(),
     '0': lambda x:x.sum()
}
dfs = df_final1.groupby([df_final1['Disbursement Date'].dt.strftime('%B, %Y'),'Source Name']).agg(aggregations)
```

```
# This is just a sample of 50 rows with the disbursement date and the source name along with the respective overdues.
dfs.head(50)
```

```
# Creating A New Column of Total Dues by adding the 30+ overdues, 60+ overdues and 90+ overdues.
dfs['Total Dues']=dfs['30+']+dfs['60+']+dfs['90+']+dfs['<30']+dfs['0']
```
